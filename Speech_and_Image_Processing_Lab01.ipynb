```python
#Importing the necessary library
import nltk
from urllib import request
from bs4 import BeautifulSoup
from nltk import word_tokenize
import re
import io

import unicodedata


import string
import sys
import argparse

import codecs
```

```python
nltk.download('punkt')
```

```python
arabic_punctuations = '''`÷×؛<>_۔()*&^%][ـ،/:"؟‘’.,'{}~¦+|!”…“–ـ'''
english_punctuations = string.punctuation
punctuations_list = arabic_punctuations + english_punctuations

arabic_diacritics = re.compile("""
							 ّ    | # Tashdid
							 َ    | # Fatha
							 ً    | # Tanwin Fath
							 ُ    | # Damma
							 ٌ    | # Tanwin Damm
							 ِ    | # Kasra
							 ٍ    | # Tanwin Kasr
							 ْ    | # Sukun
							 ـ     # Tatwil/Kashida
							’	 | # ulta comma
							‘	 | # ulta comma 2
						 """, re.VERBOSE)



def normalize_arabic(text):
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "ء", text)
    text = re.sub("ئ", "ء", text)
    text = re.sub("ة", "ه", text)
    text = re.sub("گ", "ك", text)
    return text


def remove_diacritics(text):
	text = re.sub(arabic_diacritics, '', text)
	return text


def remove_punctuations(text):
    translator = str.maketrans('', '', punctuations_list)        #used for setting to none Here, 
																																#first, the mapping between the two strings firstString and secondString are created.
																																#Then, the third argument thirdString resets the mapping of each character in it to None
																																
    return text.translate(translator)


def remove_repeating_char(text):
	return re.sub(r'(.)\1+', r'\1', text)

```

## Downloading and cleaning text from Web using BeautifulSoup

```python
url = 'https://www.bbc.com/urdu/world-54431244'
html = request.urlopen(url)
raw = BeautifulSoup(html, 'html.parser').get_text()
raw = raw.translate ({ord(c): " " for c in "\\!@#$%^&*\'\"()[]{};:,./<>?\|`~-=_+\n"})
#print(raw)

raw = remove_punctuations(raw)
#print(raw)
raw = remove_diacritics(raw)
#print(raw)
raw = remove_repeating_char(raw)
print(raw)
```

### Tokenizing the Text

```python
tokens = word_tokenize(raw)

print('The length of the words are: ' + str(len(tokens)))

tokens = [w for w in tokens if not re.match(r'[A-Z]+', w,re.I)]
#print(tokens)
tokens = [w for w in tokens if not re.match(r'[0-9]+', w, re.I)]
tokens = [w for w in tokens if not re.match(r'۔+', w, re.I)]
#print(tokens)
```
